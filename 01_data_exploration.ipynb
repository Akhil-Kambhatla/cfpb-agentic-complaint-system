```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Encode categorical features
le_product = LabelEncoder()
le_issue = LabelEncoder()

df["product_encoded"] = le_product.fit_transform(df["Product"])
df["issue_encoded"] = le_issue.fit_transform(df["Issue"])

# Feature matrix
feature_cols = [
    "product_encoded", "issue_encoded", "narrative_length",
    "mentions_regulation", "mentions_dollar_amount", 
    "mentions_dates", "mentions_attorney", "mentions_multiple_contacts"
]

X = df[feature_cols].values
y = df["got_meaningful_resolution"].values

# Train and get feature importances
model = GradientBoostingClassifier(n_estimators=100, max_depth=4, random_state=42)
scores = cross_val_score(model, X, y, cv=5, scoring="accuracy")
print(f"Cross-val accuracy: {scores.mean()*100:.1f}% (+/- {scores.std()*100:.1f}%)")

model.fit(X, y)
importances = dict(zip(feature_cols, model.feature_importances_))
print("\n=== Feature Importances ===")
for feat, imp in sorted(importances.items(), key=lambda x: -x[1]):
    print(f"  {feat}: {imp:.3f}")
```

```python
# The business framing: Risk-Adjusted Resolution Strategy
# For each complaint, estimate: risk of dismissing vs cost of resolving

# Create risk tiers based on product + features
df["regulatory_risk"] = 0.0

# High-risk products (heavily regulated, frequent CFPB enforcement)
high_risk_products = ["Debt collection", "Mortgage", 
                       "Payday loan, title loan, personal loan, or advance loan"]
medium_risk_products = ["Credit card", "Credit reporting or other personal consumer reports",
                         "Checking or savings account"]

df.loc[df["Product"].isin(high_risk_products), "regulatory_risk"] += 0.3
df.loc[df["Product"].isin(medium_risk_products), "regulatory_risk"] += 0.15
df.loc[df["mentions_regulation"] == 1, "regulatory_risk"] += 0.2
df.loc[df["mentions_attorney"] == 1, "regulatory_risk"] += 0.15
df.loc[df["mentions_multiple_contacts"] == 1, "regulatory_risk"] += 0.1

# Predict resolution probability using our trained model
df["resolution_probability"] = model.predict_proba(X)[:, 1]

# The insight: complaints with HIGH regulatory risk but LOW resolution probability
# are the dangerous ones — company is likely to dismiss but SHOULDN'T
df["risk_gap"] = df["regulatory_risk"] - df["resolution_probability"]

print("=== Risk Gap Analysis ===")
print("Complaints where company likely dismisses but regulatory risk is high:")
high_risk_dismissed = df[df["risk_gap"] > 0.2]
print(f"  Count: {len(high_risk_dismissed)} ({len(high_risk_dismissed)/len(df)*100:.1f}%)")
print(f"  Top products:")
print(high_risk_dismissed["Product"].value_counts().head())

print("\n=== Example high-risk complaint that would likely be dismissed ===")
example = high_risk_dismissed.sort_values("risk_gap", ascending=False).iloc[0]
print(f"Product: {example['Product']}")
print(f"Issue: {example['Issue']}")
print(f"Resolution probability: {example['resolution_probability']*100:.0f}%")
print(f"Regulatory risk: {example['regulatory_risk']*100:.0f}%")
print(f"Risk gap: {example['risk_gap']*100:.0f}pp")
print(f"Actual outcome: {example['Company response to consumer']}")
print(f"Narrative: {example['Consumer complaint narrative'][:400]}")
```

```python
import pymc as pm
import arviz as az
import numpy as np
import pandas as pd

# Reload data
df = pd.read_csv("data/processed/dev_set_10k.csv")
df["got_resolution"] = df["Company response to consumer"].isin([
    "Closed with monetary relief", "Closed with non-monetary relief"
]).astype(int)

# Extract features
import re
df["narrative_length"] = df["Consumer complaint narrative"].str.len()
df["mentions_regulation"] = df["Consumer complaint narrative"].str.contains(
    r'FCRA|FDCPA|TILA|RESPA|Fair Credit|Fair Debt|Regulation|Section \d|violat',
    case=False, na=False).astype(int)
df["mentions_attorney"] = df["Consumer complaint narrative"].str.contains(
    r'attorney|lawyer|legal action|sue|lawsuit|court',
    case=False, na=False).astype(int)
df["mentions_dollar"] = df["Consumer complaint narrative"].str.contains(
    r'\$[\d,]+', na=False).astype(int)

# Encode products as numeric (use resolution rate as encoding — more informative than label encoding)
product_resolution_rate = df.groupby("Product")["got_resolution"].mean()
df["product_risk_score"] = df["Product"].map(product_resolution_rate)

# Standardize continuous features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
continuous_features = ["product_risk_score", "narrative_length"]
df[["product_risk_z", "length_z"]] = scaler.fit_transform(df[continuous_features])

# Sample down for PyMC (full 10k takes too long)
df_sample = df.sample(n=2000, random_state=42)

print(f"Training Bayesian model on {len(df_sample)} complaints...")
print(f"Resolution rate in sample: {df_sample['got_resolution'].mean()*100:.1f}%")
```

```python
with pm.Model() as complaint_model:
    # Priors — weakly informative
    intercept = pm.Normal("intercept", mu=0, sigma=2)
    beta_product = pm.Normal("beta_product_risk", mu=0, sigma=2)
    beta_length = pm.Normal("beta_narrative_length", mu=0, sigma=1)
    beta_regulation = pm.Normal("beta_mentions_regulation", mu=0, sigma=1)
    beta_attorney = pm.Normal("beta_mentions_attorney", mu=0, sigma=1)
    beta_dollar = pm.Normal("beta_mentions_dollar", mu=0, sigma=1)
    
    # Linear combination
    logit_p = (
        intercept 
        + beta_product * df_sample["product_risk_z"].values
        + beta_length * df_sample["length_z"].values
        + beta_regulation * df_sample["mentions_regulation"].values
        + beta_attorney * df_sample["mentions_attorney"].values
        + beta_dollar * df_sample["mentions_dollar"].values
    )
    
    # Likelihood
    p = pm.Deterministic("p", pm.math.sigmoid(logit_p))
    outcome = pm.Bernoulli("outcome", p=p, observed=df_sample["got_resolution"].values)
    
    # Sample
    trace = pm.sample(1000, tune=500, cores=2, random_seed=42, 
                      return_inferencedata=True, progressbar=True)

print("✓ Bayesian model fitted")
print(az.summary(trace, var_names=["intercept", "beta_product_risk", "beta_narrative_length",
                                     "beta_mentions_regulation", "beta_mentions_attorney", 
                                     "beta_mentions_dollar"]))
```

```python
import pickle

# Extract posterior samples for prediction
posterior = trace.posterior
posterior_samples = {
    "intercept": posterior["intercept"].values.flatten(),
    "beta_product_risk": posterior["beta_product_risk"].values.flatten(),
    "beta_narrative_length": posterior["beta_narrative_length"].values.flatten(),
    "beta_mentions_regulation": posterior["beta_mentions_regulation"].values.flatten(),
    "beta_mentions_attorney": posterior["beta_mentions_attorney"].values.flatten(),
    "beta_mentions_dollar": posterior["beta_mentions_dollar"].values.flatten(),
}

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def predict_resolution(narrative, product, product_resolution_rates, scaler_params):
    """Predict resolution probability with uncertainty for a single complaint."""
    # Extract features
    prod_risk = product_resolution_rates.get(product, 0.4)
    narr_len = len(narrative)
    mentions_reg = int(bool(re.search(r'FCRA|FDCPA|TILA|RESPA|Fair Credit|Fair Debt|Regulation|Section \d|violat', narrative, re.I)))
    mentions_att = int(bool(re.search(r'attorney|lawyer|legal action|sue|lawsuit|court', narrative, re.I)))
    mentions_dol = int(bool(re.search(r'\$[\d,]+', narrative)))
    
    # Standardize
    prod_risk_z = (prod_risk - scaler_params["product_mean"]) / scaler_params["product_std"]
    length_z = (narr_len - scaler_params["length_mean"]) / scaler_params["length_std"]
    
    # Compute probability for each posterior sample
    logits = (
        posterior_samples["intercept"]
        + posterior_samples["beta_product_risk"] * prod_risk_z
        + posterior_samples["beta_narrative_length"] * length_z
        + posterior_samples["beta_mentions_regulation"] * mentions_reg
        + posterior_samples["beta_mentions_attorney"] * mentions_att
        + posterior_samples["beta_mentions_dollar"] * mentions_dol
    )
    probs = sigmoid(logits)
    
    return {
        "mean": float(np.mean(probs)),
        "std": float(np.std(probs)),
        "hdi_3": float(np.percentile(probs, 3)),
        "hdi_97": float(np.percentile(probs, 97)),
        "samples": probs  # full posterior for plotting
    }

# Store scaler params
scaler_params = {
    "product_mean": df[["product_risk_score"]].mean().values[0],
    "product_std": df[["product_risk_score"]].std().values[0],
    "length_mean": df[["narrative_length"]].mean().values[0],
    "length_std": df[["narrative_length"]].std().values[0],
}

# Test on our credit card example
test_narrative = "I booked a hotel through my Citi credit card travel portal. My flight got delayed and I called the hotel. They told me to contact Citi. I called twice and they promised to resolve it but did nothing. The hotel charged me $850 plus 15,000 points. I disputed and they ruled it my responsibility."

result = predict_resolution(test_narrative, "Credit card", 
                           product_resolution_rate.to_dict(), scaler_params)

print(f"=== Prediction for Credit Card complaint ===")
print(f"Resolution probability: {result['mean']*100:.1f}%")
print(f"95% Credible interval: [{result['hdi_3']*100:.1f}%, {result['hdi_97']*100:.1f}%]")
print(f"Uncertainty (std): {result['std']*100:.1f}pp")

# Now simulate intervention: what if this were treated as a Debt Collection complaint?
result_intervention = predict_resolution(test_narrative, "Debt collection",
                                          product_resolution_rate.to_dict(), scaler_params)
print(f"\n=== Intervention: If treated as Debt Collection ===")
print(f"Resolution probability: {result_intervention['mean']*100:.1f}%")
print(f"95% CI: [{result_intervention['hdi_3']*100:.1f}%, {result_intervention['hdi_97']*100:.1f}%]")
print(f"Effect of intervention: {(result_intervention['mean']-result['mean'])*100:+.1f}pp")

# Save everything for the backend
model_package = {
    "posterior_samples": posterior_samples,
    "product_resolution_rates": product_resolution_rate.to_dict(),
    "scaler_params": scaler_params,
    "feature_importances": {
        "product_type": 0.713,
        "narrative_length": -0.002,
        "mentions_regulation": -0.041,
        "mentions_attorney": -0.007,
        "mentions_dollar": 0.053,
    }
}

with open("data/processed/bayesian_model.pkl", "wb") as f:
    pickle.dump(model_package, f)

print("\n✓ Saved Bayesian model to data/processed/bayesian_model.pkl")
```
