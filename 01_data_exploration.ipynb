```python
#LLM Helper Function
import anthropic, os, json
from dotenv import load_dotenv
load_dotenv(override=True)

client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))

def ask_claude(prompt, system="You are a helpful assistant."):
    resp = client.messages.create(
        model="claude-sonnet-4-20250514",
        max_tokens=2048,
        system=system,
        messages=[{"role": "user", "content": prompt}]
    )
    return resp.content[0].text
```

```python
#Test run
system_prompt = """You are a financial complaint classifier for the CFPB (Consumer Financial Protection Bureau).

Given a consumer complaint narrative, classify it by returning ONLY valid JSON with these fields:
- predicted_product: The financial product category
- predicted_issue: The main issue
- predicted_sub_issue: The specific sub-issue (or null if unclear)
- severity: low, medium, high, or critical
- compliance_risk_score: 0.0 to 1.0 (how likely this involves a regulatory violation)
- confidence: 0.0 to 1.0
- reasoning: One sentence explaining your classification"""

prompt = f"""Classify this consumer complaint:

\"\"\"{test_complaint['Consumer complaint narrative']}\"\"\"

Respond with JSON only. No markdown fences."""

result = ask_claude(prompt, system=system_prompt)
print("RAW RESPONSE:")
print(result)

# Parse it
parsed = json.loads(result)
print("\n\nCOMPARISON:")
print(f"  Actual product:    {test_complaint['Product']}")
print(f"  Predicted product: {parsed['predicted_product']}")
print(f"  Actual issue:      {test_complaint['Issue']}")
print(f"  Predicted issue:   {parsed['predicted_issue']}")
print(f"  Severity:          {parsed['severity']}")
print(f"  Compliance risk:   {parsed['compliance_risk_score']}")
print(f"  Reasoning:         {parsed['reasoning']}")
```

```python
#Improvised classifier with contrained labels
# Get the actual CFPB labels from our dataset
products = df_work["Product"].unique().tolist()
issues_by_product = df_work.groupby("Product")["Issue"].unique().to_dict()

system_prompt_v2 = f"""You are a financial complaint classifier for the CFPB.

You MUST choose from these exact product categories:
{json.dumps(products, indent=2)}

For each product, you MUST choose from these exact issue categories:
{json.dumps({k: v.tolist() for k, v in issues_by_product.items()}, indent=2)}

Given a complaint narrative, return ONLY valid JSON:
- predicted_product: Must be one of the product categories above
- predicted_issue: Must be one of the issues for that product above
- severity: low, medium, high, or critical
- compliance_risk_score: 0.0 to 1.0
- confidence: 0.0 to 1.0
- reasoning: One sentence explaining your classification"""

prompt_v2 = f"""Classify this consumer complaint:

\"\"\"{test_complaint['Consumer complaint narrative']}\"\"\"

Respond with JSON only. No markdown fences."""

result_v2 = ask_claude(prompt_v2, system=system_prompt_v2)
parsed_v2 = json.loads(result_v2)

print("COMPARISON (v2 with constrained labels):")
print(f"  Actual product:    {test_complaint['Product']}")
print(f"  Predicted product: {parsed_v2['predicted_product']}")
print(f"  Match: {'✓' if parsed_v2['predicted_product'] == test_complaint['Product'] else '✗'}")
print(f"\n  Actual issue:      {test_complaint['Issue']}")
print(f"  Predicted issue:   {parsed_v2['predicted_issue']}")
print(f"  Match: {'✓' if parsed_v2['predicted_issue'] == test_complaint['Issue'] else '✗'}")
print(f"\n  Severity:          {parsed_v2['severity']}")
print(f"  Compliance risk:   {parsed_v2['compliance_risk_score']}")
print(f"  Reasoning:         {parsed_v2['reasoning']}")
```

```python
# Testing on 5 diff samples and Pick one from each major product type
test_indices = []
for prod in ["Credit reporting or other personal consumer reports", 
             "Debt collection", "Credit card", 
             "Checking or savings account", "Mortgage"]:
    idx = df_work[df_work["Product"] == prod].index[0]
    test_indices.append(idx)

correct_product = 0
correct_issue = 0

for idx in test_indices:
    row = df_work.loc[idx]
    prompt = f"""Classify this consumer complaint:

\"\"\"{row['Consumer complaint narrative'][:1500]}\"\"\"

Respond with JSON only. No markdown fences."""

    try:
        result = ask_claude(prompt, system=system_prompt_v2)
        parsed = json.loads(result)
        
        p_match = parsed["predicted_product"] == row["Product"]
        i_match = parsed["predicted_issue"] == row["Issue"]
        correct_product += p_match
        correct_issue += i_match
        
        print(f"Product: {'✓' if p_match else '✗'}  Issue: {'✓' if i_match else '✗'}  | {row['Product'][:30]}")
        if not p_match:
            print(f"  predicted: {parsed['predicted_product']}")
        if not i_match:
            print(f"  predicted issue: {parsed['predicted_issue']}")
            print(f"  actual issue:    {row['Issue']}")
    except Exception as e:
        print(f"Error on {row['Product']}: {e}")

print(f"\nProduct accuracy: {correct_product}/5")
print(f"Issue accuracy:   {correct_issue}/5")
```
